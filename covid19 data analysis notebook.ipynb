## Import packages

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
print('hello')
```

### Import and transform the dataset 

```python
data_origin=pd.read_csv("Covid19_Confirmed_dataset.csv")
data_origin.head(10)
```

```python
data_origin.shape
```

### Delete irrelevant columns

```python
data_transform=data_origin.drop(['Lat','Long'],axis=1)
data_transform
```

### group by country & region

```python
df_cov=data_transform.groupby('Country/Region').sum()
```

```python
df_cov.head(10)
```

```python
df_cov.shape
```

### Visualize specific countries to see the trend

```python
df_cov.loc['China'].plot()
df_cov.loc['Italy'].plot()
df_cov.loc['US'].plot()
plt.legend()
```

```python
df_cov.loc['China'][:3].plot()
```

### Find measurement of the intensity of virus spreading

```python
#the diff of infection
df_cov.loc['China'].diff().plot()
df_cov.loc['Italy'].diff().plot()
df_cov.loc['US'].diff().plot()
plt.legend()
```

```python
#the sum of infection 
df_cov['infect_sum'] = df_cov.apply(lambda x: x.sum(),axis=1)
df_cov.sort_values(by='infect_sum', ascending=False)
```

```python
# the max diff of infection for all countries
countries=list(df_cov.index.values)
countries
max_infect=[]

for c in countries:
    max_infect.append(df_cov.loc[c].diff().max())

df_cov['max_infect']=max_infect
```

```python
df_cov.head()
```

### Make new dataframe

```python
df1=pd.DataFrame(df_cov[['max_infect','infect_sum']])
df1.head()
```

### Read new datasets

```python
#coutry data

df_countinfo=pd.read_csv('worldwide_happiness_report.csv')
df_death=pd.read_csv('covid19_deaths_dataset.csv')

df_death.head()
```

```python
df_countinfo.head()
df_countinfo.set_index('Country or region',inplace=True)
```

```python
df_c=df_countinfo.drop(['Overall rank','Perceptions of corruption','Generosity','Score'],axis=1)
df_c.head()
```

```python
df_death.drop(['Lat','Long'],axis=1,inplace=True)  
df_d=df_death.groupby('Country/Region').sum() 
```

```python
df_d['death_sum']=df_d.apply(lambda x: x.sum(),axis=1)
df_d.head()
```

```python
dfd=pd.DataFrame(df_d['death_sum'])
dfd.head()
```

### Join the three datasets

```python
dfnew=df1.join(df_c,how='inner').join(dfd,how='inner') 
```

```python
dfnew.head()
```

### Find the correlation

```python
dfnew.corr()
```

```python

x=dfnew['GDP per capita']
y=dfnew['max_infect']
```

### Transform y because of different metrics

```python
sns.scatterplot(x,np.log(y))
sns.regplot(x,np.log(y))
```

### Plotting Social support vs maximum Infection rate

```python
x=dfnew['Social support']
y=np.log(dfnew['max_infect'])

with sns.axes_style('white'):
    sns.jointplot(x, y, data=dfnew, kind='hex')
```

### Plotting Freedom to make life choices vs maximum Infection rate

```python
x=dfnew['Freedom to make life choices']
y=dfnew['infect_sum']

sns.regplot(x,np.log(y))
```
